```python
DF_1['Classification'].value_counts().plot(kind='barh', figsize=(8, 6))
plt.xlabel("Count", labelpad=14)
plt.ylabel("classification of iris", labelpad=14)
plt.title("Iris Classification", y=1.02);
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1, 2, 1)
sns.countplot(DF_1['SL'])
plt.title('Distribution of Sepel Length')
plt.xticks(rotation = 90)


plt.subplot(1, 2, 2)
sns.countplot(DF_1['SW'])
plt.title('Distribution of Sepel Width')
plt.xticks(rotation = 90) 
plt.show()
```

Here sepel lenth 5.0 is the highest count and sepel width 3.0 is the highest count.

```python
plt.figure(figsize = (15,6))

plt.subplot(1,2,1)
sns.countplot(DF_1['PL'])
plt.title('Distribution of Petal Length')
plt.xticks(rotation = 90) 
plt.show()

plt.figure(figsize = (15,6))
plt.subplot(1,2,2)
sns.countplot(DF_1['PW'])
plt.title('Distribution of Petal Width')
plt.xticks(rotation = 90) 
plt.show()
```

Petal length 1.5 is the highest count and petal width 0.2 is the highest count
